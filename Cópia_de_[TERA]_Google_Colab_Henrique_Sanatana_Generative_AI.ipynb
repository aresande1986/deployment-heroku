## Importando as bibliotecas

```python
import pandas as pd
from transformers import pipeline
```

## Dados

```python
feedbacks_df = pd.read_csv('data.csv')
```

```python
feedbacks_df.head()
```

## Modelo

```python
model_path = "citizenlab/twitter-xlm-roberta-base-sentiment-finetunned"
sentiment_classifier = pipeline("text-classification", model=model_path, tokenizer=model_path, max_length=512, truncation=True)
```

## Funções

```python
def classify(text:str):
    result = sentiment_classifier(text)[0]
    result['input_text'] = text
    return result
```

```python
def classify_df(df:pd.DataFrame, review_column:str):
    classified_data = df[review_column].apply(classify).tolist()
    result_df = pd.DataFrame(classified_data)
    return result_df
```

## Classificando um review qualquer

```python
classify("Poderia ter sido muito melhor. Mas tudo bem")
```

## Classificando o primeiro

```python
classify(feedbacks_df['review'][0])
```

## Classificando todos

```python
classify_df(feedbacks_df, 'review')
```
